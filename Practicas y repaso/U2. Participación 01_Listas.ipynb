# Ejercicio 1

Promedio de ventas  
Una tienda registra las ventas diarias en una lista:  
      ventas = [120, 150, 90, 200, 175, 80, 300]  
 a) Calcular el promedio de ventas.  
 b) Mostrar cuántos días estuvieron por encima del promedio.  
 Caso de prueba 1: usar las ventas ejemplo  
 Caso de prueba 2: leer por teclado el número de ventas y el rango de valores de las ventas;   
                   luego, generar aleatoriamente la lista de ventas  


```python
#caso de prueba 1
ventas = [120, 150, 90, 200, 175, 80, 300]
```

```python
#caso de pruba 2
n= int(input("Introduce número ventas:"))
mini= float(input("Introduce el mínimo de ventas:"))
maxi= float(input("Introduce el máximo de ventas:"))


import random
ventas = [random.uniform(mini,maxi) for _ in range (n)]
print("Ventas:", ventas)
```

```python
#caso de prueba 1
ventas = [120, 150, 90, 200, 175, 80, 300]

#caso de pruba 2
n= int(input("Introduce número ventas:"))
mini= float(input("Introduce el mínimo de ventas:"))
maxi= float(input("Introduce el máximo de ventas:"))


import random
ventas = [random.uniform(mini,maxi) for _ in range (n)]
print("Ventas:", ventas)


# (a)
prom = sum (ventas)/len(ventas)
print(f"Promedio de ventas {prom}")

# (b) Mostrar cúantos días estuvieron por encima del promedio
acc = 0
for v in ventas:
    if v > prom:
        acc = acc + 1 
print(f"Número de días por encima del promedio: {acc}")
```

# Ejercicio 2

Detección de valores atípicos simples  
Se registran temperaturas:  
    temperaturas = [20, 22, 19, 21, 35, 20, 23, 18]  
a) Calcular la media.  
b) Mostrar qué valores están 10 unidades por encima o por debajo de la media (posibles outliers).  
Caso de prueba 1: usar las temperaturas ejemplo  
Caso de prueba 2:   
* leer por teclado el número de temperaturas
* el rango de valores de las temperaturas;
* luego, generar aleatoriamente la lista de temperaturas  

```python
# caso d prueba 1
temperatura = [20, 22, 19, 21, 35, 20, 23, 18] 
```

```python
# caso de prueba 2
n= int(input("Introduce las temperaturas:"))
mini= float(input("Introduce el mínimo de temperatura:"))
maxi= float(input("Introduce el máximo de temperatura:"))


import random
temperatura = [random.uniform(mini,maxi) for _ in range (n)]
print("Temperaturas:", temperatura)
```

```python
# a) calcular la media 
prom = sum(temperatura)/len(temperatura)
print(f"El promedio es: {prom}")
```

```python
# b)Mostrar que valores están 10 unisades por encima o por debajo de la media(posibles outliers)
outliers = [t for t in temperatura if t > prom+10 or t < prom-10]
print(outliers)
```
