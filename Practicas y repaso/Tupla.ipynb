# Tupla

Una tupla es una estructura de datos se denotra con `

Características:  
* Es ordenada: mantiene el orden de inserción de sus valores, sus valores tienen posiciones.
* Es imutable: sus valores no pueden cambiarse despúes de haber creado la estructura, su tamaño no cambia.
* Permite duplicados: permite repetir valores.
* Es heterogénea: los valores pueden ser de diferente tipo.
* Es it6erable. puede regresar sus valores uno a uno.

Una tupla es más eficiente que una lista, ado que sus valores no cambian, el acceso a ellos es más eficiente, tanto en memoria como en tiempo de ejecución.

Una tupla se usa cuando se agrupan datos que "naturalmente" siempre van juntos. Por ejemplo la fecha de nacimiento.





## Crear una tupla

```python
# crear directamente con sus valores

vocales= ('a', 'e', 'i', 'o', 'u')
vocales, type(vocales)
```

```python
# crear con otras estructuras de datos mediante un casting 
vocales = tuple("aeiou" )
vocales, type(vocales)
```

```python
# crear tupla con un solo valor
anio = (2004,) # si necesitamos una tupla de un solo valor se necesita poner una coma ,
anio, type(anio)
```

```python
# crear una tupla vacia
vacia = ()
vacia, type(vacia)
```

```python
# crear sun usar paréntesis, solamente con sus valores
# separado por comas
# EMPAQUETAR
vocales= 'a','e','i','o','u'
vocales, type(vocales)
```

```python
# DESEMPAQUETAR
vocales= 'a','e','i','o','u'
v1,v2,v3,v4,v5 = vocales
v1,v2,v3,v4,v5, type(v1), type(v2)
```

```python
# crear con tuple-comprenhesion
# crea una tupla con multi´plos de 5 en el rago de [5,50]

tupla = tuple(n*5 for n in range (1, 11))
tupla, type(tupla)
```

```python
# crear con tuple-comprenhesion
# crea una tupla con multi´plos de 5 en el rago de [5,50]
tupla = tuple(n for n in range (5,51,5))
tupla, type(tupla)
```

```python
# crear tupla con tuple comprenhesion 
# de las consonantes minúsculas

import string

string.ascii_lowercase
consonantes = tuple(x for x in string.ascii_lowercase if x not in ('aeiou'))
print(consonantes)
```

## Acceso a valores



```python
# dada una posición, calcualr el valor de esa posicion
vocales = tuple("aeiou")
vocales [2]
print(vocales[2])

# dado un valor calcular la posición

vocales.index('i')
```

## Recorrer

```python
tupla = tuple(n*5 for n in range (1, 11))

#recorrer por posición (mediante índice (posiciones de valores))
for i in range (len(tupla)):
    print(tupla[i])

#recorrer por valor
for v in tupla:
    print(v)
```
