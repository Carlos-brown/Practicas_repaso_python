# Repaso. Sentencia Repetitiva while

`for` se usa cuando se conoce el úmero de iteraciones a realizar
`while` se usa cuando no se conoce el número de iteraciones a realiza

Sintaxis:

```python
while condicion:
    bloque
```

`condidicion` debe ser booleano. Sib la `condicion` es verdadera, se 
ejecuta el `bloque` y regresa a verificar la `condidicon`. Si es falsa,
termina el ciclo

````python
# Ejemplo: sumar el total de ventas de un día. Se leen de teclado las 
# ventas hasta ingresar 0
# Técnica 1: leer 1er valor antes de ciclo
# leer siguiente valores al terminar el ciclo

acc = 0
venta= float(input("Ingresa venta: $")) #lee la primera venta

while venta != 0:
    acc = acc + venta
    venta= float(input("Ingresa venta: $")) # lee siguiente venta
print(f"Venta total: ${acc}")
````

````python
# Ejemplo: sumar el total de ventas de un día. Se leen de teclado las 
# ventas hasta ingresar 0
# Técnica 2: usar una bandera

acc = 0
esCero = False

while not esCero: #operador not n lugar de ! #si la primera es false y se vuelve a negar, entonces se vuelve positivo y entra
    venta= float(input("Ingresa venta: $"))
    if venta == 0:
        esCero= True
    else:
        acc = acc + venta
print(f"Venta total: ${acc}")
````
