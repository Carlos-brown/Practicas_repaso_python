## Decisión

Sentencia desiciva de un cambio en Python

```python 
if condición:
    bloque
```

`condición` debe ser una expresión cuya evaluación debe ser un 
booleano.

Cuando `condición` es verdadera, se eejecuta bloque. Cuando es falsa, 
no se ejecuta.

````python
# Dado los coeficientes de una ecuación de grado 2,
# decidir si las raíces son reales

a = 8 #3
b = 2 #8
c = 3 #2

disc = b**2-4*a*c
if disc > 0:
    print("raíces reales")
````

Sentencia desiciva de dos caminos en Python

```python 
if condición:
    bloque
else:
    bloque2
```

`condición` debe ser una expresión cuya evaluación debe ser un 
booleano.

Cuando `condición` es verdadera, se ejecuta `bloque1` y `bloque2`, no. Cuando es falsa, no se ejecuta `bloque1`, se ejecuta `bloque2`.

````python
# Dado los coeficientes de una ecuación de grado 2,
# decidir si las raíces son reales o no. 
# Si son reales, calcularlas

a = 3#8 #3
b = 8#2 #8
c = 2#3 #2

disc = b**2-4*a*c
if disc > 0:
    print("raíces reales")
    x1= (-b + disc**.5) / (2*a)
    x2= (-b - disc**.5) / (2*a)
    print(x1,x2)
else:
    ("Raíces no reales")
````

Sentencia desiciva de tres caminos en Python

```python 
if condición1:

    if (condición2):

```

que se puede escribir como sigue:
```python
if condición1:
```

`condición1`, `condición2` deben ser expresiones cuya evaluación debe ser un 
booleana.

Cuando `condición1` es verdadera, se ejecuta `bloque1` y `bloque2` y `bloque3`, no. Cuando es falsa, evalúa `condición2`.

Cuando `condición2` es verdadera, se ejecuta `bloque2` y `bloque3`, no. Cuando es falsa, se ejecuta `bloque3`.

````python
# Dado los coeficientes de una ecuación de grado 2,
# decidir si las raíces son reales o no. 
# Si son reales, decidir si son iguales o no, y calcularlas

a = 1 #8 #3
b = 6 #2 #8
c = 9 #3 #2

disc = b**2-4*a*c
if disc > 0:
    print("raíces reales diferentes")
    x1 = (-b + disc**.5) / (2*a)
    x2 = (-b - disc**.5) / (2*a)
    print(x1,x2)
elif disc == 0:
    x = -b/(2*a)
    print("Raíces reales son iguales")
    print(x)
else:
    ("Raíces no reales")
````

## Repetición for 

Sintaxis de for controlado por contador

```python
for i in range(n):
    bloque
```

````python
for i in range(3): #contar del 0 al 2
    print(i)
for i in range(1,4): #contar del 1 al 3
    print(i)
for i in range(2,7,2): #tercer argumento son los pasos, contar del 2 al 6 en 2 en 2
    print (i)
for i in range(0,7,2): 
    print (i)
for i in range(6,1,-1): 
    print (i)
    
````

## Ejemplo de usos de for

# Leer n enteros de teclado y calcular promedios

````python
n = int(input("Ingresa n:"))
acc = 0
for i in range (n):
    numero = int(input("Ingrese número: "))
    acc = acc + numero
prom = acc/n
print("prom= ", prom)
````

````python
# Cálculo de máxio de un conjunto de n enteros
# que representan edades de personas

n = int(input("Ingresa n: "))
maxi = -1
for i in range(n):
    edad = int(input("Ingrese edad:"))
    if edad > maxi:
        maxi = edad
print("edad máxima:",maxi)
````
