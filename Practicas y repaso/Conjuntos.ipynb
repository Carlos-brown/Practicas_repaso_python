# Conjuntos

Un conjunto es una estructura de datos con ls siguientes características:
* No ordenado: no mantiene el orden de inserción. Los valores no mantiene posiciones.
* Es mutable: sus valores pueden modificarse despúes de su creación.
* No permite duplicados: todos sus valoes son únicos.
* Es heterogéneo: cada valor puede ser diferente tipo de datos.
* Es iterable: puede refresar sus valores uno a uno

Se denota con `{}` y el tipo de dato es `set`

# Crear conjuntos

```python
# crear conjunto directamente con sus valores

vocales = {'a', 'e', 'i', 'o', 'u'}
print(vocales)
print(type(vocales))
```

```python
# crear conjunto utilizando otra estructura de datos

vocales = set("aeiou")
print(vocales)
print(type(vocales))
```

```python
# crear conjunto vacío

vacio = {} #esto es un diccionario vacio
vacio = set ()
print(vacio)
print(type(vacio))
```

```python
# crear conjunto utilizando otra estructura de datos

vocales = set("aeiou")
binario = set[(0,1)]
print(vocales)
print(type(vocales))
print(binario)
print(type(binario))
```

# Acceder a valores

como los valores de un conjunto no tienen posiciones, la única forma de acceder es mediante su valor en un for

```python

vocales = {'a', 'e', 'i', 'o', 'u'}
for v in vocales:
    print(v)
```

# Insertar valores

```python
# insertar con add un solo valor
vocales = {'a', 'e', 'i', 'o', 'u', 'e'}
vocales.add('A')
print(vocales)
```

```python
# insertar con update muchos valores

vocales = {'a', 'e', 'i', 'o', 'u'}
vocales.update({'A','E','I','O','U'})# ['A','E','I','O','U']'AEIOU')
print(vocales)
```

# Eliminar valores

```python
# eliminar con respecto al valor con remove
vocales = {'a', 'e', 'i', 'o', 'u'}
vocales.remove('o')
# vocles.remove('z') # Marca error
print(vocales)
```

```python
# eliminar con respecto al valor con discard
vocales = {'a', 'e', 'i', 'o', 'u'}
vocales.discard('o')
vocales.discard('z')
print(vocales)
```

```python
# eliminar un valor con pop y regres el valor eliminado

vocales = {'a', 'e', 'i', 'o', 'u'}
v = vocales.pop()
print(v)
print(vocales)
```

```python
# eliminar todos los valores. genera conjunto vacío, con clear

vocales = {'a', 'e', 'i', 'o', 'u'}
vocales.clear()
print(vocales)
```
