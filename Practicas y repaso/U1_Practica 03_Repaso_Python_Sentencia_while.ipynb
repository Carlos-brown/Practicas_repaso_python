# Ejercicio 1
#### Comparación con referencia. Producto más cercano a un precio objetivo.
#### Dado por teclado un precio objetivo y una serie de precios de productos hasta ingresar 0,
#### por cada producto: calcula la diferencia absoluta entre el precio y el precio objetivo
#### y determina cuál producto es el más cercano al objetivo. 
#### Mostrar nombre del producto y su precio

```python
precioObje = float(input("Ingrese el precio objetivo del producto"))

menor_diferencia = 0
producto_cercano = ""
precio_cercano = 0
primero = True

while True:
    producto = input("Ingrese el nombre del producto: ")
    precio = float(input("Ingrese precio del producto: $"))

    if precio == 0:
        break

        
    dif_abs = precio - precioObje

    if dif_abs < 0:
        dif_abs = dif_abs * -1
    if primero:
        menor_diferencia = dif_abs
        producto_cercano = producto
        precio_cercano = precio
        primero = False
    else:
        if dif_abs < menor_diferencia:
            menor_diferencia = dif_abs
            producto_cercano = producto
            precio_cercano = precio
    


print(F"El producto más cerca al objetivo es {producto_cercano} con un precio de {menor_diferencia}")

```

# Ejercicio 2
#### Control de calidad de datos. Conteo de precios fuera de rango aceptable
#### Dado por teclado una serie de precios hasta ingresar 0,
#### y considerando que un precio es válido si está entre $50 y $1000,
#### e inválido en caso contrario, calcular cuántos precios son válidos
#### y cuántos inválidos

```python
precio= float(input("Ingrese el precio: $ "))
              
precio_V = 0
precio_NV = 0

while precio  != 0:
    precio= float(input("Ingrese el precio: $ "))
    
    if precio >= 50 and precio <= 1000:    
        precio_V = precio_V + 1
    else:
            precio_NV = precio_NV + 1

print("Los precios válidos son: ", precio_V)
print("Los precios inválidos son:", precio_NV)
        
```

# Ejercicio 3
#### Variabilidad local. Mayor diferencia entre dos valores consecutivos
#### Leer números hasta que se ingrese -1.
#### Determinar cuál es la mayor diferencia absoluta entre dos valores consecutivos.

```python
valor_ant =float(input("Ingrese un número"))

mayor_diferencia= 0.0

if valor_ant != -1:
    valor_act= float(input("Ingrese el siguiente número: "))

    while valor_act != -1:
        dif = valor_act - valor_ant

        if dif < 0:
            dif = dif * -1
        if dif > mayor_diferencia:
            mayor_diferencia = dif

        valor_ant = valor_act
        valor_act = float(input("Ingrese el siguiente número: "))
        
if mayor_diferencia == 0 and valor_ant == -1:
    print("No se ingresaron suficientes datos")
else:
    print("La mayor diferencia absoluta es: ", mayor_diferencia)

print("valor actual", valor_act)
```

# Ejercicio 4
#### Comportamiento de una serie. Conteo de cambios de signo
#### Leer números hasta que se ingrese 999.
#### Contar cuántas veces la secuencia cambia de: positivo a negativo, negativo a positivo

```python
n= float(input("Ingrese un número"))

cambio_PN = 0
cambio_NP = 0
cambio_total =0

while n != 999:
    num = float(input("Ingrese un número"))

    if num == 999:
        break

        
    if n  > 0 and num < 0:
        cambio_PN = cambio_PN + 1
        cambio_total = cambio_total + 1

    if n < 0 and num > 0:
        cambio_NP = cambio_NP + 1
        cambio_total = cambio_total + 1

    n = num
    
print(f"La sencuencia cambio de positivo a negativo {cambio_PN} veces")
print(f"La sencuencia cambio de negativo a positivo {cambio_NP} veces")
print(f"Los cambios totales son: {cambio_total} veces")
```

# Ejercicio 5
#### Detección de patrón. Longitud de la racha positiva más larga
#### Leer números hasta que se ingrese -1.
#### Determinar cuál fue la mayor cantidad de números positivos consecutivos.

```python
n = float(input("Ingrese un número: "))

positivo = 0
mayor_cant = 0

while n != -1:
    if n > 0:
        positivo = positivo + 1
        if positivo > mayor_cant:
            mayor_cant = positivo
    else:
        positivo = 0

    n = float (input("Ingrese un número: "))

print(f"La mayor cantidad de números positivos consecutivos fue {mayor_cant}")
```
