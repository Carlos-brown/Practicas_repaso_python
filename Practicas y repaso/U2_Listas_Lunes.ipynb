# Listas

# una lista es ua estructura de datos nativa de python que se caracteriza por:

* Es ordenada: mantiene el orden de inserción, cada valor tiene a posición en la estructura

* Permite duplicados: se pudeden repetir valores

* es heterogenea: los valores puden ser de diferente tipo

* Es iterable; puede regresar sus valores uno a uno

# Crear listas

```python
# Crear lista directamente con sus valores
alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
print (alumno)
print(type(alumno))
```

```python
# Crear lista vacía}
l = []
l
```

```python
# crear lista con otas estructuras de datos

l = list("hola")
l
```

```python
# crear lista con list-comprehension (describiendo sus valores)
# crear una lista que tenga 5 veces la cadena "hola"

l = ["hola" for i in range(5)]
l
```

```python
# crear lista con los valores del 5 al 15 con list-comprehesion

l = [x for x in range(5,16)]
l
```

```python
# crear lista con pares con pares enteros entere 4 y 20 con list-comprehesion}

l = [x for x in range(4,21,2)]
l = [x for x in range(4,21) if x%2==0]
l = [2*x for x in range(2,11)]
l
```

```python
# crear lista con potencia de 2  en el rango [2, 4096]

l = [2**x for x in range (2, 13)]
l
```

```python
# crear lista con potencia de 2  en el rango [2, 4096]
# logaritmo
import math

l = [2**x for x in range (1, int(math.log2(4096)+1))]
l
```

```python
# crear una lista del alfabeto en minúsculas
import string
l = list(string.ascii_lowercase)
print(l)
```

```python
# crear lista de consonantes minúsculascon list-comprehesion

# l = [x for x in range(4,21) if x%2==0]
# l = [x for x in string.ascii_lowercase if x not in ['a','e','i','o','u']]
l = [x for x in string.ascii_lowercase if x not in 'aeiou']
print (l)
```

```python
# crear una lista a partir de dos listas con operador de concatenación 
import string
l1 = list(string.octdigits)
l2 = list(string.punctuation)
l3 = l1+ l2
print (l3)
```

### Acceder a valores 


```python
# acceso a un valor por posición

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
alumno [2]
```

```python
#acceso a posición por valor

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
alumno.index("Nombre")
```

```python
#Acceso al valor de la última posición

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
alumno [6]
alumno [-1] #última posicion
alumno [-4]
alumno[len(alumno)-1]
```

```python
# acceso a sublistas(rebanadas, slices)

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
rebanada = alumno[3:6]
rebanada
```

```python
# accede a las calificaciones para sumarles un punto

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
alumno[3:6] = [9.7,10,7.4]

alumno[3:6] = [alumno [i] + 1 if alumno[i]+1<=10  else alumno[i] for i in range(3,6)]

alumno
```

# Insertar nuevos valores


```python
alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
alumno.append(20)
alumno
```

```python
# insertar con extend al final de la lista muchos valores

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
alumno.extend(["TICs",20,"PAD"])
alumno
```

```python
# agregar en una posición un valor con insert

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
alumno.insert(1,20)
alumno
```

# Eliminar valores

```python
# eliminar un valor con remove

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
alumno.remove("Nombre")
alumno
```

```python
# eliminar el valor de la }ultima posición

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
valor = alumno.pop()
valor, alumno
```

```python
# eliminar el valor de una posición con del

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
del alumno[2]
alumno
```

```python
# elimnar rebanadaas con del
# elimnar con calificaciones

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]

del alumno[3:6]
alumno
```

```python
# elimina la lista 

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
del alumno
alumno
```

```python
# vaciar la lista con clear

alumno = ["22220932","Nombre",8,8.7,9,6.4, True]
alumno.clear()
alumno
```
