# Ejercicio 1 
#### Conteo: distrbución básica
#### Dado por teclado un conjunto de n calificaciones en la escala 0 a 10,
#### calcular número de aprobadas y número de no aprobadas,
#### siendo la calificacíon más alta 

```python
n = int(input("Ingrese número de calificaciones: "))

a = 0
na = 0

for i in range(n):
    calificacion = float(input("Ingresa la calificación:"))
    if calificacion < 0 or calificacion  > 10:
        print("calificación invalida")
    else:
        if calificacion >= 6:
            a = a + 1
        else:
            na = na + 1
print("total de aprobadas: ", a)
print("total de no aprobadas:", na)
```

# Ejercicio 2 
#### Promedio: media de tendencia central 
#### Dado por teclado un conjunto de n calificaciones en la escala 0 a 10,
#### calcular el promedio global, el promedio de aprobados y 
#### el promedio de no aprobadas

```python
n = int(input("Ingrese número de calificaciones:"))

prom_a = 0
prom_na = 0
prom_global = 0

a = 0
na =0 
total = 0

for i in range(n):
    calificacion = float(input("Ingrese la calificación:"))
    if calificacion < 0 or calificacion > 10:
        print ("Calificación invalida")
    else:
        prom_global = prom_global + calificacion
        total = total + 1
        if calificacion >=6:
            prom_a = prom_a + calificacion
            a = a + 1
        else:
            prom_na = prom_na + calificacion
            na = na + 1
            
if total > 0 :
    promedio_g = prom_global / n
else:
    promedio_g = 0
if  a > 0:
    promedio_a = prom_a / a
else:
    promedio_a = 0
if na > 0:
    promedio_na = prom_na / na
else:
    promedio_na = 0
    
print("Promedio de aprobadas=", promedio_a)
print("Promedio de no aprobadas=", promedio_na)
print("Promedio global=", promedio_g)
```

# Ejercicio 3
#### Máximos y mínimos: Valores extremos
#### Dado por teclado un conjunto de n calificaciones den la escala de 0 al 10
#### calcular la calificacion máxima y mínima

```python
n = int (input("Ingrese número de calificaciones"))

calif_max=-1
calif_min= 11

for i in range(n):
    calificacion = float(input("Ingrese la calificación: "))
    if calificacion <0 or calificacion>10:
        print("Calificación invalida")
    else:
        if calificacion > calif_max:
            calif_max = calificacion
        if calificacion < calif_min:
            calif_min = calificacion

print("Calificación máxima= ", calif_max)
print("Calificación mínima= ",calif_min)
```

# Ejercicio 4 
#### Rango: dispersión básica 
#### Dado por el teclado un conjunto de n calificaciones en la escala de 0 al 10
#### calcular el rango (rango = máximo- mínimo)

```python
n = int (input("Ingrese número de calificaciones"))

calif_max=-1
calif_min= 11

for i in range(n):
    calificacion = float(input("Ingrese la calificación: "))
    if calificacion <0 or calificacion>10:
        print("Calificación invalida")
    else:
        if calificacion > calif_max:
            calif_max = calificacion
        if calificacion < calif_min:
            calif_min = calificacion

rango = calif_max - calif_min

print("Calificación máxima= ", calif_max)
print("Calificación mínima= ",calif_min)
print("Rango= ", rango)
```

# Ejercicio 5 
#### Porcentaje: proporciones
#### Dado por teclado un conjunto de n calificaciones en la escala 0 al 10
#### calcular el procentaje de aprobadas, no aprobadas y calificaciones de 10

```python
n = int(input("Ingrese número de calificaciones:"))

porc_a = 0
porc_na = 0
calif_diez = 0

for i in range(n):
    calificacion = float(input("Ingrese la calificación:"))
    if calificacion < 0 or calificacion > 10:
        print ("Calificación invalida")
    else:
        if calificacion >=6:
            porc_a = porc_a + 1
            
        else:
            porc_na = porc_na + 1

        if calificacion == 10:
            calif_diez = calif_diez +1
            
porcentaje_g = (calif_diez / n) * 100
porcentaje_a = (porc_a / n) * 100
porcentaje_na = (porc_na / n) * 100
    
print("Porcentaje de aprobadas=", porcentaje_a, "%")
print("Porcentaje de no aprobadas=", porcentaje_na, "%")
print("Porcentaje de calificaciones de 10=", porcentaje_g, "%")
```
